```python
#Create DataFrame
# Load dataset into dataframe

df = pd.read_csv('Dataset_for_CA1.csv')

#Print first few rows of the dataset
print(df.head())

# data information
df.info()
```

```python
#Obtain the leading digits for every value

# Get the leading digit by defining a function
def find_leading_digit(value):
    return int(str(value)[0])

# Use the function on 'A-D' column and generate a new column named 'Leading Digits'
df['Leading_Digit A'] = df['A'].apply(find_leading_digit)
df['Leading_Digit B'] = df['B'].apply(find_leading_digit)
df['Leading_Digit C'] = df['C'].apply(find_leading_digit)
df['Leading_Digit D'] = df['D'].apply(find_leading_digit)


# Print the stored values for 'Leading Digits of column A-D'  in a DataFrame
print(df[['Country (or dependency)', 'A', 'Leading_Digit A']])
print(df[['Country (or dependency)', 'B', 'Leading_Digit B']])
print(df[['Country (or dependency)', 'C', 'Leading_Digit C']])
print(df[['Country (or dependency)', 'D', 'Leading_Digit D']])
```

```python
#Find leading digits as a percentage of the total.

#Determine the total count of each leading digit
leading_digit_counts_A = df['Leading_Digit A'].value_counts().sort_index()

leading_digit_counts_B = df['Leading_Digit B'].value_counts().sort_index()

leading_digit_counts_C = df['Leading_Digit C'].value_counts().sort_index()

leading_digit_counts_D = df['Leading_Digit D'].value_counts().sort_index()


# Determine each leading digit count's percentage in relation to the overall count.
overall_count = len(df)

percentage_of_total_A = (leading_digit_counts_A / overall_count) * 100

percentage_of_total_B = (leading_digit_counts_B / overall_count) * 100

percentage_of_total_C = (leading_digit_counts_C / overall_count) * 100

percentage_of_total_D = (leading_digit_counts_D / overall_count) * 100

# Print the results of columns A-D's leading digit counts.
print("Leading Digit Counts:")
print(leading_digit_counts_A)
print(leading_digit_counts_B)
print(leading_digit_counts_C)
print(leading_digit_counts_D)


# Print the percentage of column A-D
print("\nPercentage of Total:")
print(percentage_of_total_A)
print(percentage_of_total_B)
print(percentage_of_total_C)
print(percentage_of_total_D)
```

```python
#Use Matplotlib to plot percentages vs. leading digits in a bar graph with the proper labelling

import matplotlib.pyplot as plt
#import numpy as np

# Generate a 2x2 grid of subplots
fig, axes = plt.subplots(2, 2, figsize=(12, 8))

# Load Data into the DatFrame
df = pd.read_csv('Dataset_for_CA1.csv')

# Generate a list of these datasets
datasets_benford = [percentage_of_total_A, percentage_of_total_B, percentage_of_total_C, percentage_of_total_D]

# Subplot labelling
subplot_labelling = ['A', 'B', 'C', 'D']

# Using the loop condition for iterating through the subplots and datasets
for i, ax in enumerate(axes.ravel()):
    ax.bar(range(1, len(datasets_benford[i]) + 1), datasets_benford[i])
    ax.set_xlabel(f'Leading Digits_{subplot_labelling[i]}')
    ax.set_ylabel(f'Percentage of Total_{subplot_labelling[i]} (%)')
    ax.set_title(f'Percentage of Leading Digits in Column {subplot_labelling[i]}')

# Change the arrangement to avoid subplot overlap.
plt.tight_layout()

# Show the plot
plt.show()
```

```python
#Insert the leading digits calculation into its own function

import pandas as pd
import matplotlib.pyplot as plt

# determine the leading digits by defining a function
def determine_leading_digits(data, ColumnName):

    # Extract leading digit for the specified column
    def find_leading_digit(value):
        return int(str(value)[0])

    # Generate a new column for leading digits
    data['Leading_Digit ' + ColumnName ] = data[ColumnName].apply(find_leading_digit)

    # Determine leading digit counts
    leading_digit_counts = data['Leading_Digit ' + ColumnName].value_counts().sort_index()

   # Determine each leading digit count's percentage in relation to the overall count.
    overall_count = len(data)
    percentage_of_total = (leading_digit_counts / overall_count) * 100

    return leading_digit_counts, percentage_of_total

# Determine leading digits for column 'A -D' using the function
leading_digit_counts_A, percentage_of_total_A = determine_leading_digits(df, 'A')
leading_digit_counts_B, percentage_of_total_B = determine_leading_digits(df, 'B')
leading_digit_counts_C, percentage_of_total_C = determine_leading_digits(df, 'C')
leading_digit_counts_D, percentage_of_total_D = determine_leading_digits(df, 'D')

# Print the results of columns A-D's leading digit counts.
print("Leading Digit Counts:")
print(leading_digit_counts_A)
print(leading_digit_counts_B)
print(leading_digit_counts_C)
print(leading_digit_counts_D)

# Print the percentage of column A-D
print("\nPercentage of Total:")
print(percentage_of_total_A)
print(percentage_of_total_B)
print(percentage_of_total_C)
print(percentage_of_total_D)
```

```python
#Compare the values found to those expected with Benford’s law

import pandas as pd
import matplotlib.pyplot as plt

#  Determine leading digits by defining the function
def Determine_leading_digits(data, ColumnName):

    # Derive leading digit for the specified column
    def find_leading_digit(value):
        return int(str(value)[0])

    # Generate a new column for leading digits
    data['Leading_Digit'] = data[ColumnName].apply(find_leading_digit)

    # Get leading digit counts
    leading_digit_counts = data['Leading_Digit'].value_counts().sort_index()

    # Determine each leading digit count's percentage in relation to the overall count.
    overall_count = len(data)
    percentage_of_total = (leading_digit_counts / overall_count) * 100

    return percentage_of_total

# Load dataset into a DataFrame
df = pd.read_csv('Dataset_for_CA1.csv')

# Generate a figure for subplots
fig, axes = plt.subplots(2, 2, figsize=(12, 8))

fig.suptitle("Comparison of Leading Digit Frequencies")

columns_benford = ['A', 'B', 'C', 'D']

for i, column in enumerate(columns_benford):
    examined_frequencies = Determine_leading_digits(df, column)

    predicted_frequencies = [30.1, 17.6, 12.5, 9.7, 7.9, 6.7, 5.8, 5.1, 4.6]

    # Generate a bar plot for found and predicted frequencies
    axes[i // 2, i % 2].bar(range(1, 10), examined_frequencies)
    axes[i // 2, i % 2].plot(range(1, 10), predicted_frequencies, marker='o', linestyle='--', color='red')
    axes[i // 2, i % 2].set_title(f'Column {column}')
    axes[i // 2, i % 2].set_xlabel('Leading Digit')
    axes[i // 2, i % 2].set_ylabel('Percentage of Total (%)')
    axes[i // 2, i % 2].set_xticks(range(1, 10))

   # Add a legend to the subplot
    axes[i // 2, i % 2].legend(["predicted frequncies", "examined frequencies"])


plt.tight_layout()
plt.subplots_adjust(top=0.9)
plt.show()
```
